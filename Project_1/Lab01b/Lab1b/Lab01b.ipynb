### 2a. Plot the data in a single plot

Using the [Matplotlib](https://matplotlib.org/) documentation, visualize the dataset with Matplotlib. All charts should include a title, axis labels and a legend when applicable.

1. Import 'lab1.csv' into a ndarray using genfromtxt. 
2. Set figure size to 12x8. Use either [plt.figure](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html) or [rcParams](https://matplotlib.org/3.1.0/tutorials/introductory/customizing.html#matplotlib-rcparams) (*search for `figure.figsize`*).

```python
import matplotlib.pyplot as plt

csv = np.genfromtxt('lab1.csv', delimiter=",")

csv2 = csv.reshape(-1, 1)

#before making plot
plt.figure(figsize =(12,8))

x = csv[:, - 1]
y = csv[:, 0]
#make a line plot 
plt.plot(x, y, 'r--', label='Data')

# plot it 
# plt.plot(x,y) .. , color="r"..., linewidth=5


plt.title('Cartesian Line Graph')

plt.xlabel('X Axis')

plt.ylabel('Y Axis')

plt.legend()

plt.show()


#plt.xlim() and ylim

```

### 2b. Plot the data in multiple subplots

Create 6 graphs in 2 columns and 3 rows using [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html). 
1. Row 1 is two subplots (a scatterplot and a line graph) of the 2nd column graphed on the x-axis VS the 1st column graphed on the y-axis
2. Row 2 is two subplots (a scatterplot and a line graph) of the 3rd column (x-axis) VS the 2nd column (y-axis)
3. Row 3 is two subplots (a scatterplot and a line graph) of the 3rd column (x-axis) VS the 1st column (y-axis)
4. Include a title and axis labels for each subplot. You can refer to the 1st column as `X`, the 2nd as `Y`, and the 3rd as `Z`.
5. To insure there is enough space between subplots use [subplots_adjust](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots_adjust.html)

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(5, 10))

x1 = csv[:,1]
y1 = csv[:,0]

x2 = csv[:,2]
y2 = csv[:,1]


axes[0,0].plot(x2, y2, color=plt.cm.tab10.colors[0], linewidth=5, linestyle='dashed', label='log(x)')
axes[0,0].set_title('P1')
axes[0,0].set_xlabel('x values')
axes[0,0].set_ylabel('y values')
axes[0,0].set_xlim(-20, 20)
axes[0,0].set_ylim(-6, 10)
axes[0,0].legend()


axes[0,1].scatter(x2, y2)
axes[0,1].set_title('S1')
axes[0,1].set_xlabel('x values')
axes[0,1].set_ylabel('y values')
axes[0,1].set_xlim(-20, 20)
axes[0,1].set_ylim(-10, 10)



axes[1,0].scatter(x1, y1)
axes[1,0].set_title('S2')
axes[1,0].set_xlabel('x values')
axes[1,0].set_ylabel('y values')
axes[1,0].set_xlim(-10, 10)
axes[1,0].set_ylim(0, 5)



axes[1,1].plot(x1, y1, color=plt.cm.tab10.colors[0], linewidth=2, linestyle='dashed', label='log(x)')
axes[1,1].set_title('P2')
axes[1,1].set_xlabel('x values')
axes[1,1].set_ylabel('y values')
axes[1,1].set_xlim(-10, 10)
axes[1,1].set_ylim(0, 6)
axes[1,1].legend()
plt.subplots_adjust(hspace=1)

axes[2,0].scatter(x2, y1)
axes[2,0].set_title('S3')
axes[2,0].set_xlabel('x values')
axes[2,0].set_ylabel('y values')
axes[2,0].set_xlim(-20, 20)
axes[2,0].set_ylim(-1, 5)
plt.subplots_adjust(hspace=1)



axes[2,1].plot(x2, y1, color=plt.cm.tab10.colors[0], linewidth=2, linestyle='dashed', label='log(x)')
axes[2,1].set_title('P3')
axes[2,1].set_xlabel('x values')
axes[2,1].set_ylabel('y values')
axes[2,1].set_xlim(-10, 10)
axes[2,1].set_ylim(0, 10)
axes[2,1].legend()
plt.subplots_adjust(hspace=1)
```
